Visualicemos las variables em cuanto a posibles relaciones

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Configuración de estilo para los gráficos
sns.set(style="whitegrid")
```

```python
# 1. Pairplot para visualizar las relaciones entre todas las variables
sns.pairplot(df, hue='target', palette='Set2', diag_kind='kde')
plt.suptitle("Relaciones entre variables (Iris dataset)", y=1.02)
plt.show()
```

```python
# 2. Mapa de calor de correlación
correlation = df.drop(columns=["target"]).corr()  # Excluir la columna 'target' para el cálculo
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Mapa de calor de correlaciones entre características")
plt.show()
```

```python
# 3. Gráficos de distribución de cada característica por clase (target)
for column in df.columns[:-1]:  # Itera sobre todas las columnas menos 'target'
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=df, x=column, hue="target", fill=True, common_norm=False, alpha=0.5, palette="Set2")
    plt.title(f"Distribución de {column} por clase")
    plt.xlabel(column)
    plt.ylabel("Densidad")
    plt.show()
```

```python
# 4. Boxplot para observar la dispersión de las características por clase
plt.figure(figsize=(10, 6))
df_melted = df.melt(id_vars='target', var_name='Característica', value_name='Valor')
sns.boxplot(data=df_melted, x='Característica', y='Valor', hue='target', palette='Set2')
plt.title("Dispersión de características por clase (target)")
plt.xticks(rotation=45)
plt.legend(title='Clase')
plt.show()
```

```python
# 5. Gráficos de dispersión seleccionados (ejemplo: petal width vs. petal length)
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df, x="petal length (cm)", y="petal width (cm)", hue="target", palette="Set2", s=100)
plt.title("Relación entre petal length y petal width por clase")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title='Clase')
# Guardar el gráfico como archivo
output_path = "scatter_petal_length_vs_width.png"  # Cambia la extensión a .jpg, .pdf, etc., si lo prefieres
plt.savefig(output_path, dpi=300, bbox_inches="tight")  # dpi para calidad, bbox_inches para ajustar márgenes
plt.show()
```

### ¿Cómo están distribuídos los datos dentro de la muestra respecto de las especies?. En caso de tener más ejemplos de un tipo de flor, y menos de otro tipo ¿crees que influiría en el algoritmo final? ¿por qué?

```python
especies = df['target'].value_counts()

print(especies)

etiquetas = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
especies.index = especies.index.map(etiquetas)


plt.figure(figsize=(8, 6))
sns.barplot(x=especies.index, y=especies.values, palette="Set2")
plt.title("Distribución de especies en el dataset Iris")
plt.xlabel("Especie")
plt.ylabel("Cantidad")
plt.show()
```

La distribución de las especies en el dataset Iris es equilibrada con 50 registros. Este balance evita posibles sesgos en el modelo y permite que el algoritmo tenga un rendimiento similar para todas las clases. Si existiese un desbalance, sería necesario aplicar técnicas como sobremuestreo para garantizar que las clases minoritarias no queden subrepresentadas.
